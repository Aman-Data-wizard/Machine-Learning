```python
import pandas as pd
```

```python
data = pd.read_csv('/content/CSv_ ML.csv')
data.head()
```

data.describe()

```python
data['Salary'] = data['Salary'].fillna(data['Salary'].mean())
data['CGPA'] = data['CGPA'].fillna(data['CGPA'].mean())
data.head()
```

```python
cgpa = data[['CGPA']]
salary = data['Salary']
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(cgpa,salary)
result = model.predict([[10]])
print("Predicted salary", result)
```

```python
#PCA feature reduction technique
import numpy as np
```

```python
fish = np.array([[200,100,5],[190,90,5], [100,90,3],[90,70,3]])
type = np.array([0,0,1,1])  #0 salmon #1 catfish
fish
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
fish_scaler  = scaler.fit_transform(fish)
pca = PCA(n_components=3)
fish_pca = pca.fit_transform(fish_scaler)
print("PCA: ", pca.explained_variance_ratio_)
print("fish_pca: ", fish_pca)
```
